# Hitters Dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/rahulbanerjee/Rahul/ETL_HIVE/ETL_Hive_Assignments/Pandas/hitters.csv')
```

```python
df.head()
```

```python
# Find out no of missing values in each column
df.isna().sum()
```

```python
#Find of people who have batted for more than 400 matches
df.iloc[np.where(df.AtBat > 400)]['Name']
```

```python
#Find out the name of the player who has hit max home runs

#Option 1
df[df['HmRun']== max(df['HmRun'])]

#Option 2
df.sort_values(by=['HmRun'],ascending=False).head(1)['Name']
```

```python
# Find out league wise total number of players
df.groupby(['League']).agg({'Name':'size'})
```

```python
#Find out division wise total amount of salary invested
df.groupby(['Division']).agg({'Salary':'sum'})
```

```python
#Find out five most senior players. 
df.sort_values(by=['Years'],ascending=False).head(5)[['Name','Years']]
```

# Cars Dataset

```python
df1= pd.read_csv('/Users/rahulbanerjee/Rahul/ETL_HIVE/SQL/Cars93.csv')
```

```python
#Find out car having highest Mileage in city
df1.sort_values(by=['MPG.city'],ascending=False).head(1)
```

```python
#Find out Models of cars which are having Price between 17 and 25
df1[(df1['Price'] > 17) & (df1['Price'] < 25) ][['Model','Price']]
```

```python
# Find out top 7 most fuel economic cars for highway
df1.sort_values(by=['MPG.highway'],ascending=False).head(7)
```

```python
# Find out Models of cars having no AirBags and which are of compact type whose price is below 20.
df1[(df1['AirBags'] == 'None') & (df1['Type'] == 'Compact') & (df1['Price'] < 20)][['Model']]
```

```python
# Find out top 7 most fuel economic cars for city, show results in sorted manner based on Horsepower.
cars_df = df1.sort_values(by=['MPG.city'],ascending=False).head(7)
cars_df.sort_values(by=['Horsepower'],ascending=False)[['Manufacturer','Model','Horsepower']]
```

```python
# Find out Cars having mileage of more than 15 miles in city
df1[df1['MPG.city'] > 15]
```

```python
#Find out Cars with US based origin
df1[df1['Origin'] == 'USA']
df1.iloc[np.where(df1['Origin'] == 'USA')]
```

```python
#Find out Cars with no airbags and which are compact in type, from the output retrieve only Model, Mileage in city, Type, Airbags
df1[(df1['AirBags'] == 'None') & (df1['Type'] == 'Compact') ][['Model','MPG.city','Type']]
```

```python
#Find out cars with Mileage in city more than 15 which are compact in type.
df1[(df1['MPG.city'] > 15) & (df1['Type'] == 'Compact') ]
```
